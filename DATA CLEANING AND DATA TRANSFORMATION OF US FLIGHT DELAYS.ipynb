# Data Understanding, Data Cleaning, and Data Transformation

Show how to:
- Understand the data
- Identify missing values
- Handle missing values
- Remove unused variables

Role distribution: This dataset contains Flight Delay data from January 2017 to July 2022.


```python
#import data
import pandas as pd
file= 'C:/Users/david/OneDrive/Documents/JEFIN/Portfolio Airline Delay/Airline_Delay_Cause.csv'
df = pd.read_csv(file)
```

```python
#data
df.head(10)
```

```python
#info data
df.info()
```

```python
#drop unused variable
data=df.drop(["carrier", "arr_flights", "arr_del15", "carrier_ct", "weather_ct", "nas_ct", "security_ct", "late_aircraft_ct", "arr_cancelled", "arr_diverted"], axis=1)
data
```

```python
#missing value
data.isna().sum()
```

```python
#fill na arr_delay using mean
mean_arr_delay = data['arr_delay'].mean()
data['arr_delay'] = data['arr_delay'].fillna(mean_arr_delay)

#fill na carrier_delay using mean
mean_carrier_delay = data['carrier_delay'].mean()
data['carrier_delay'] = data['carrier_delay'].fillna(mean_carrier_delay)

#fill na weather_delay using mean
mean_weather_delay = data['weather_delay'].mean()
data['weather_delay'] = data['weather_delay'].fillna(mean_weather_delay)

#fill na nas_delay using mean
mean_nas_delay = data['nas_delay'].mean()
data['nas_delay'] = data['nas_delay'].fillna(mean_nas_delay)

#fill na security_delay using mean
mean_security_delay = data['security_delay'].mean()
data['security_delay'] = data['security_delay'].fillna(mean_security_delay)

#fill na late_aircraft_delay using mean
mean_late_aircraft_delay = data['late_aircraft_delay'].mean()
data['late_aircraft_delay'] = data['late_aircraft_delay'].fillna(mean_late_aircraft_delay)
```

```python
#missing value
data.isna().sum()
```

```python
data['day']=1
data
```

```python
#data transformation (date)
data['Date']=pd.to_datetime(data[['year','month','day']])
data
```

```python
#data transformation (city and airport name)
data[['city', 'Airport_name']] = data['airport_name'].str.split(':\ ', expand=True)
data['Airport_name'] = data['Airport_name'].str.rstrip(':')
```

```python
#data transformation (city and state)
data[['city', 'state']] = data['city'].str.split(',\ ', expand=True)
data['state'] = data['state'].str.rstrip(',')
data
```

```python
#drop unused variable
datafinal=data.drop(["year", "month", "day", "airport_name"], axis=1)
```
